#Import dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

#Scrap Wikipedia data

```python
wiki_data = pd.read_html("https://en.m.wikipedia.org/wiki/List_of_countries_by_minimum_wage#")
```

#EDA

```python
display(wiki_data[2].head())
my_table = wiki_data[2]

#to select top 10 countries by minimum wage
my_table = my_table.iloc[:, [0,2]]
display(my_table)
```

```python
my_table.dropna()
```

```python
my_table.columns=["Country", "Min_Wage"]
```

```python
my_table = my_table.sort_values("Min_Wage", ascending=False).head(10)
```

```python
my_table
```

#Plot

```python
my_table = my_table.sort_values("Min_Wage", ascending=True)
```

```python
def get_color(name, number):
    pal = list(sns.color_palette(palette=name, n_colors=number).as_hex())
    return pal
```

```python
pal_vi = get_color('viridis_r', len(my_table))
```

```python

import math
plt.gcf().set_size_inches(12, 12)
sns.set_style('darkgrid')

#set max value
max_val = max(my_table['Min_Wage'])*1.01
fig, ax = plt.subplots()
ax = plt.subplot(projection='polar')

#set the subplot 
ax.set_theta_zero_location('N')
ax.set_theta_direction(1)
ax.set_rlabel_position(0)
ax.set_thetagrids([], labels=[])
ax.set_rgrids(range(len(my_table)),labels= my_table['Country'])

#set the projection
ax = plt.subplot(projection='polar')

for i in range(len(my_table)):
    ax.barh(i, list(my_table["Min_Wage"])[i]*1.65*np.pi/max_val,
            label=list(my_table["Country"])[i], color=pal_vi[i])

plt.title("Top 10 countries by annual minimum wage in USD")
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()
```
