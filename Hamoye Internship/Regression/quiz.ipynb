```python
# Question 12
x, y = df[['T2']], df['T6']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
from sklearn.metrics import r2_score
round(r2_score(y_test, linear_model.predict(x_test)), 2)
```

```python
# Question 13
from sklearn.preprocessing import MinMaxScaler
df_copy = df.drop(columns=['date', 'lights'])
scaler = MinMaxScaler()
norm_df = pd.DataFrame(scaler.fit_transform(df_copy), columns=df_copy.columns)
x = norm_df.drop(columns=['Appliances'])
y = norm_df['Appliances']
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, test_size=0.3, random_state=42)
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
from sklearn.metrics import mean_absolute_error
round(mean_absolute_error(y_test, linear_model.predict(x_test)), 2)
```

```python
# Question 14
from sklearn.metrics import mean_absolute_error
t  = mean_squared_error(y_test, linear_model.predict(x_test))
error = t * x_test.shape[0]
round(error, 2)
```
